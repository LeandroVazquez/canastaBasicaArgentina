Origen de datos: https://datos.gob.ar/pt_BR/dataset/sspm-canasta-basica-alimentaria-regiones-pais/archivo/sspm_445.1

```python
## Cargo las librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Cargo el archivo que usaré para realziar las gráficas
canastaBasica=pd.read_excel("canasta-basica-total-regiones-del-pais.xlsx")
```

```python
##Separamos la columna fecha en tres, día, mes y año
canastaBasicaNueva=canastaBasica
canastaBasicaNueva["DiaMesAño"]=canastaBasicaNueva['indice_tiempo'].astype(str)
canastaBasicaNueva['Año']=canastaBasicaNueva["DiaMesAño"].str[0:4]
canastaBasicaNueva['Mes']=canastaBasicaNueva["DiaMesAño"].str[5:7]
canastaBasicaNueva['Día']=canastaBasicaNueva["DiaMesAño"].str[8:10]
canastaBasicaNueva
```

```python
## Genero el gráfico con la variación del monto de la canasta básica, separada por región
canastaGranBA=canastaBasicaNueva.groupby("Año")["gran_buenos_aires"].mean() 
canastaCuyo=canastaBasicaNueva.groupby("Año")["cuyo"].mean()
canastaNorEste=canastaBasicaNueva.groupby("Año")["noreste"].mean()
canastaNorOeste=canastaBasicaNueva.groupby("Año")["noroeste"].mean()
canastaPampeana=canastaBasicaNueva.groupby("Año")["pampeana"].mean()
canastaPatagonia=canastaBasicaNueva.groupby("Año")["patagonia"].mean()

index=['2017','2018','2019','2020','2021']
tablaGeneral=pd.DataFrame({'Gran Buenos Aires':canastaGranBA,'Cuyo':canastaCuyo,'NorEste':canastaNorEste,'NorOeste':canastaNorOeste,'Pampeana':canastaPampeana,'Patagonia':canastaPatagonia},index=index)
tablaGeneral.plot(kind='line')
plt.title("Variación del monto de la canasta básica desde el año 2017 al 2021, separada por región")
plt.xlabel("Años")
plt.ylabel("Monto en $")
plt.show()
```
